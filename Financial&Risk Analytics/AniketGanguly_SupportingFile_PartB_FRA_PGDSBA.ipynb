```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv('Market_Risk_Data_coded.csv', parse_dates=['Date'], index_col='Date')
```

```python
print(data.head())
print()
data.isnull().sum()
```

```python
# Plotting stock prices over time
plt.figure(figsize=(12, 8))

for column in data.columns:
    plt.plot(data.index, data[column], label=column)

plt.title('Stock Price vs Time')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate returns
returns = data.pct_change().dropna()

# Calculate mean and standard deviation of returns
mean_returns = returns.mean()
std_dev_returns = returns.std()

# Plot Mean vs. Standard Deviation
plt.figure(figsize=(10, 6))
plt.scatter(std_dev_returns, mean_returns, alpha=0.5)
plt.xlabel('Standard Deviation of Returns')
plt.ylabel('Mean of Returns')
plt.title('Mean vs. Standard Deviation of Returns')
for i, txt in enumerate(returns.columns):
    plt.annotate(txt, (std_dev_returns[i], mean_returns[i]))
plt.grid(True)
plt.show()
```
